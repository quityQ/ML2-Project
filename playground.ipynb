```python
import ollama as ol

model = 'llama3'
```

```python
# Simple try and except block to catch an error if the model is not found

try:
    response = ol.chat(model, messages=[
        {
            'role': 'user',
            'content': 'What is the best way to play Sand King in Dota 2?'
        },
    ])
    print(response['message']['content'])
except ol.ResponseError as e:
    print('Error:', e.error)
    ol.pull(model)
```
