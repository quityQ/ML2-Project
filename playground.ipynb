```python
#Basic usage of local model with ollama
import ollama as ol

model = 'llama3'

def chat(input):
    try:
        response = ol.chat(model, messages=[
            {
                'role': 'user',
                'content': input
            },
        ])
        return response['message']['content']
    except ol.ResponseError as e:
        print('Error:', e.error)
        ol.pull(model)


print(chat('what is langchain'))
```

```python
#SCraping the "Tips" section from howdoiplay using selenium
from selenium import webdriver
from selenium.webdriver.common.by import By

URL = 'https://howdoiplay.com/'

driver = webdriver.Chrome()

driver.implicitly_wait(1)

driver.get(URL+ '?' + 'Anti-Mage')

tips = driver.find_element(by=By.CLASS_NAME, value='tips')

content = tips.text

driver.quit()

print(content)
```
